```python
import nltk, random
from nltk.corpus import movie_reviews
from nltk.probability import FreqDist
```

```python
# extracting all words from moview review
all_words = FreqDist(w.lower() for w in movie_reviews.words())
len(all_words)
```

```python
word_features = list(all_words)[:2000]
```

```python
word_features
```

```python
documents = [(list(movie_reviews.words(fileid)), category)
             for category in movie_reviews.categories()
             for fileid in movie_reviews.fileids(category)]
```

```python
len(documents)
```

```python
documents[0]
```

```python
random.seed = 202409110500
random.shuffle(documents)
```

```python
def document_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features[f"contains({word})"] = (word in document_words)
    
    return features
```

```python
document_features(movie_reviews.words("pos/cv000_29590.txt"))
```

```python
featuresets = [(document_features(d), c) for d, c in documents]
```

```python
train_set, test_set = featuresets[100:], featuresets[:100]
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
nltk.classify.accuracy(classifier, test_set)
```

```python
classifier.show_most_informative_features(5)
```

```python
predicted = []
actual = []
```

```python
for contains, category in test_set:
    pred = classifier.classify(contains)
    predicted.append(pred)
    actual.append(category)
```

```python
TP = 0
FP = 0
FN = 0
```

```python
for pred, act in zip(predicted, actual):
    if pred == 'pos' and act == "pos":
        TP += 1
    elif pred == "pos" and act == 'neg':
        FP += 1
    elif pred == "neg" and act == "pos":
        FN += 1
```

```python
precision_pos = TP / (TP + FP)
recall_pos = TP / (TP + FN)
```

```python
print(f'Precision for pos: {precision_pos}')
print(f'Recall for pos: {recall_pos}')
```

```python
print(nltk.ConfusionMatrix(test=predicted, reference=actual))
```
